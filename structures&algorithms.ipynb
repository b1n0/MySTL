```python
def compare(a,b):
    if a>b: return 1
    elif a==b: return -1
    else: return 2

def merge(a, b, cmp = compare):
    if len(a) == 0:
        return b
    elif len(b) == 0:
        return b
    res = []; i = 0; j = 0
    while i < len(a) and j < len(b):
        if cmp(a[i],b[j]):            #____BIGGER____
            res.append(a[i]); i+=1
        else:
            res.append(b[j]); j+=1
    while i < len(a): res.append(a[i]); i+=1
    while j < len(b): res.append(b[j]); j+=1
    
    return res
    
def median(m):
    return sum(m)/len(m)

def sort(arr, median = median, cmp = compare):
    if len(arr) == 0 or len(arr) == 1 : return arr
    a = []; b = []; k = 0
    m = median(arr)
    for i in arr:
        if cmp(i,m)==2 : a.append(i)
        elif cmp(i,m)==-1 and k%2 == 0: a.append(i)
        else: b.append(i)
        k += 1
    del(arr)
    return merge(sort(a), sort(b))
```

```python
def bin_search(m, val): ### WHILE TRUE
    if len(m) == 0: return -1
    n = len(m)//2
    if val > m[len(m)-1]: return -1
    if val < m[0]: return -1
    while True:
        if m[n] > val:
            n = n - n//2
        elif m[n] < val:
            n = n + n//2
        else: return n
```

```python
def right_child(m, i): return (2*i+2 if 2*i+2 < len(m) else -1)
def left_child(m, i): return (2*i+1 if 2*i+1 < len(m) else -1)

def sift_down(m, i):
    if m == []: return
    left = left_child(m, i)
    right = right_child(m, i)
    imax = i
    if left > 0 and m[left] > m[imax]:
        imax = left
    if right > 0 and m[right] > m[imax]:
        imax = right
    if imax != i:
        tmp = m[i]; m[i] = m[imax]; m[imax] = tmp
        sift_down(m, imax)
    
def heapify(m):
    n = len(m)
    if n == 0: return m
    for i in range(n//2 -1,-1, -1):
        sift_down(m, i)
    return m

def pop_max(m):
    n = len(m)
    if n > 0:
        mx = m[0]
        m[0] = m[len(m)-1]
        m.pop(len(m) - 1)
        if n-1 > 0:
            sift_down(m,0)
        return mx

def heapsort(m):
    n = len(m); res = []  
    if n > 0:
        heapify(m)
        for i in range(n):
            res.append(pop_max(m))
    return res

heapsort([3,2,2,1,2,1,3])
```

```python
class stack:
    
    def __init__(self):
        self.l = []; self.len = 0; self.m = []
        
    def push(self, val):
        if self.is_emty(): 
            self.m.append(val)
        else: 
            if val > self.m[self.len - 1]: 
                self.m.append(val)
            else: 
                self.m.append(self.m[self.len - 1])
        self.l.append(val)
        self.len += 1 
        
    def pop(self):
        if self.is_emty(): return None
        self.len-=1 
        self.m.pop(self.len)
        return self.l.pop(self.len)
    
    def is_emty(self): return self.len == 0
    
    def get_top(self): 
        if self.is_emty(): return None
        return self.l[self.len - 1]
    
    def get_max(self): 
        if self.is_emty(): return None
        return self.m[self.len - 1]
    
    @property
    def length(self): return self.len
```

```python
class queue:
    def __init__(self):
        self.ls = stack(); self.rs = stack()
        self.len = 0
    
    def push(self, val):
        self.ls.push(val)
        self.len+=1
        
    def pop(self):
        if self.len == 0: return None
        self.len -= 1
        if self.rs.is_emty() == False:
            return self.rs.pop()
        while not self.ls.is_emty():
            self.rs.push(self.ls.pop())
        return self.rs.pop()

    @property
    def length(self): return self.len
```
